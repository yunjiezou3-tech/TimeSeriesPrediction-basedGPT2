## Zero-Shot M4-->M3

### Yearly

```python

import os
import random
import numpy as np
import torch
from exp.exp_zero_shot_forecasting import Exp_Zero_Shot_Forecast
from utils.tools import dotdict
fix_seed = 2021
random.seed(fix_seed)
torch.manual_seed(fix_seed)
np.random.seed(fix_seed)

args = dotdict()
args.task_name = 'zero_shot_forecast'
args.root_path = './dataset/tsf/'
args.model_id='m4_Yearly'
args.model = 'AutoTimes_Gpt2'
args.test_data_path = 'm3_yearly_dataset.tsf'
args.data = 'tsf'
args.is_training = 0
args.features = 'M'
args.seq_len = 12 # length of input（train） series
args.label_len = 6 # length of prefix of decoder（decoder 用的前缀长度）
args.token_len = 6 # 模型每次预测长度（AutoTimes 的预测是 token_len 为窗口的多步滚动预测）
args.test_seq_len = 12
args.test_label_len = 6
args.test_pred_len = 6 # test_pred_len 可能不是 token_len 的整数倍 → 需要多轮推理
args.batch_size = 16
args.learning_rate = 0.0001
args.mlp_hidden_layers = 2
args.train_epochs = 10
args.use_amp = False
args.cosine = False
args.tmax = 10
args.mix_embeds = False
args.drop_last = False
args.checkpoints = './checkpoints/'
args.val_set_shuffle = True
args.seasonal_patterns = 'Yearly'
args.dropout = 0.1
args.llm_ckp_dir = './gpt2'
args.mlp_hidden_dim = 512
args.mlp_activation = 'tanh'
args.num_workers = 10
args.patience = 3
args.des = 'test'
args.loss = 'SMAPE'
args.lradj = 'type1'
args.weight_decay = 0.00001
args.test_dir = 'short_term_forecast_m4_Yearly_AutoTimes_Gpt2_slm4_ll12_tl6_lr6_bt0.0001_wd16_hd1e-05_hl512_cos2_mixFalse_False'
args.test_file_name = 'checkpoint.pth'
args.gpu = 0
args.visualize = True
print('Args in experiment:')
print(args)
```

```python

os.environ["CUDA_VISIBLE_DEVICES"] = '0'
exp = Exp_Zero_Shot_Forecast(args)
# setting record of experiments
setting = '{}_{}_{}_sl{}_ll{}_tl{}_lr{}_bt{}_wd{}_hd{}_hl{}_cos{}_mix{}_{}'.format(
    args.task_name,
    args.model_id,
    args.model,
    args.data,
    args.seq_len,
    args.label_len,
    args.token_len,
    args.learning_rate,
    args.batch_size,
    args.weight_decay,
    args.mlp_hidden_dim,
    args.mlp_hidden_layers,
    args.cosine,
    args.mix_embeds,
    args.des)
# print('>>>>>>>start training : {}>>>>>>>>>>>>>>>>>>>>>>>>>>'.format(setting))
# exp.train(setting)
print('>>>>>>>testing : {}<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<'.format(setting))
exp.test(setting, test=True)
torch.cuda.empty_cache()
```
